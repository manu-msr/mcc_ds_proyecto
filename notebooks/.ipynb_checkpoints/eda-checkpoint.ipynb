### Conexión con la Base de Datos

```python
import psycopg2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Parámetros de conexión
param_dic = {
    "host"      : "localhost",
    "database"  : "cancerdb",
    "user"      : "manu",
    "password"  : "foogi"
}

# Conexión e importanción de datos desde la base.
conn = psycopg2.connect(**param_dic)
df = pd.read_sql_query('SELECT * FROM esophageal_clean',con=conn)
```

```python
df.mode().head(1)[['region', 'age', 'sex', 'alcohol','tobacco','khat','pain_swallowing','weight_loss','cough','status_patient']]
```

```python
# Crosstab de personas que fuman y toman (causas)
crosstab1 = pd.crosstab(df['status_patient'],[df['khat'],df['alcohol'],df['tobacco']],margins=True,margins_name='total')
crosstab1
```

```python
fig, axes = plt.subplots(4, 2, figsize=(10, 8), sharex=True, sharey=True)

sns.barplot(x=pd.Series(['Sobrevivieron','Fallecieron']), y=crosstab1.loc[:,(0,0,0)][0:2], ax=axes[0, 0])
sns.barplot(x=pd.Series(['Sobrevivieron','Fallecieron']), y=crosstab1.loc[:,(0,0,1)][0:2], ax=axes[0, 1])
sns.barplot(x=pd.Series(['Sobrevivieron','Fallecieron']), y=crosstab1.loc[:,(0,1,0)][0:2], ax=axes[1, 0])
sns.barplot(x=pd.Series(['Sobrevivieron','Fallecieron']), y=crosstab1.loc[:,(0,1,1)][0:2], ax=axes[1, 1])
sns.barplot(x=pd.Series(['Sobrevivieron','Fallecieron']), y=crosstab1.loc[:,(1,0,0)][0:2], ax=axes[2, 0])
sns.barplot(x=pd.Series(['Sobrevivieron','Fallecieron']), y=crosstab1.loc[:,(1,0,1)][0:2], ax=axes[2, 1])
sns.barplot(x=pd.Series(['Sobrevivieron','Fallecieron']), y=crosstab1.loc[:,(1,1,0)][0:2], ax=axes[3, 0])
sns.barplot(x=pd.Series(['Sobrevivieron','Fallecieron']), y=crosstab1.loc[:,(1,1,1)][0:2], ax=axes[3, 1])

axes[0, 0].set(xlabel='', ylabel='', title='No khat / No alcohol / No tabaco')
axes[0, 1].set(xlabel='', ylabel='', title='No khat / No alcohol / Sí tabaco')
axes[1, 0].set(xlabel='', ylabel='', title='No khat / Sí alcohol / No tabaco')
axes[1, 1].set(xlabel='', ylabel='', title='No khat / Sí alcohol / Sí tabaco')
axes[2, 0].set(xlabel='', ylabel='', title='Sí khat / No alcohol / No tabaco')
axes[2, 1].set(xlabel='', ylabel='', title='Sí khat / No alcohol / Sí tabaco')
axes[3, 0].set(xlabel='', ylabel='', title='Sí khat / Sí alcohol / No tabaco')
axes[3, 1].set(xlabel='', ylabel='', title='Sí khat / Sí alcohol / Sí tabaco')


fig.suptitle('Tabla de supervivencia separada por causas', fontsize=15);
```

## Principal Component Analysis (PCA)

```python
import psycopg2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
```

```python
# Parámetros de conexión
param_dic = {
    "host"      : "localhost",
    "database"  : "cancerdb",
    "user"      : "manu",
    "password"  : "foo"
}

# Conexión e importanción de datos desde la base.
conn = psycopg2.connect(**param_dic)
df = pd.read_sql_query('SELECT * FROM esophageal_clean',con=conn)
df
```

```python
X = df.drop(columns=['status_patient','id_esophageal','region'])
y = df['status_patient']
```

```python
scaler = MinMaxScaler()
data_rescaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components = 0.95)
pca.fit(data_rescaled)
reduced = pca.transform(data_rescaled)
ndf = pd.DataFrame(reduced)
ndf
```

```python
ndf = pd.concat([ndf, y], axis = 1)
ndf
```

```python
ndf.to_csv('pca.csv')
```
